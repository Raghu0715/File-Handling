```python
x=np.array([10,20,40,84,30])
y=np.array([15,25,65,22,36])
plt.plot(x,y,"p-.r", ms=20,mec="b",mfc="y")# marker color, #marker edge color, marker
plt.show()

"""plt.plot(x, y, "p-.r", ...)

"p": Specifies the marker shape as a pentagon.
"-.": Specifies the line style as dash-dot.
"r": Specifies the color of the line as red.
Customizations:

ms=20 (Marker Size): Sets the size of the markers to 20.
mec="b" (Marker Edge Color): Sets the edge color of the markers to blue.
mfc="y" (Marker Face Color): Sets the fill color of the markers to yellow."""
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data={
    'month': ['january', 'feb', "march", "april", "may","june", "july", "aug", "sep", "Oct", "Nov", "Dec"],
    "product A":[100 ,110, 130, 150, 170, 160, 140, 120, 130,140, 150, 160],
    "product B":[150,140,120,180,190,120,160,170,125,130,190,160],
    "product C":[120,140,160,180,110,130,150,170,140,190,150,100]
      }

df=pd.DataFrame(data)
df.set_index("month", inplace=True)
print(df)
```

```python
#plot the heat map
plt.figure(figsize=(12,8))
sns.heatmap(df, annot=True, fmt="d",cmap="YlGnBu", linewidth=0.5, linecolor="gray")
plt.title("montly sales data for product A, B, C", fontsize=16)
plt.xlabel("products", fontsize=12)
plt.ylabel("month", fontsize=12)
plt.show()
```
