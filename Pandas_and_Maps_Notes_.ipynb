```python
#map() method:-by using this method it tranforms the values in series using specific mapping data function
#maps takes only dictonaries and series
#map():- rewrites the old series in to the new series
#syntax: new_sr=old.sr.map(mapping)
data={"Grades":["A","B","C","D","E"]}
df=pd.DataFrame(data)
grades={'A':'Excellent',"B":"Good","C":"Fair","D":"Pass","E":"Fail"}
df['Grades']=df['Grades'].map(grades)
print(df)
#NOTE:- in apply menthod it takes function and runs overs the data frame and in map method takes new informationn or mapping verdata
```

***TIME SERIES***


```python
#what is time series data?
""" it is a sequence of data points in cronological
order that is used by business to analyize the data and make the future prediction"""
#ex:-measuring the weather temperature
#ex:- no of taxi rides in a month
#predicting a company stock price for next day
```

```python
#components of time series data
#it consists of 4 components
#1.Trend component:- this is a varaiation that moves up or down in a reasonably area                                      patteong peroid
#2.Seaonality component:- the varations that is regular and perodically and repeat it self overspecifc period
#3.Cyclic Components:- this is the varation that corresponds with business or economic "boom-bust" cycles or it follows its own peculiar cycles or irs follows its own pecular cycles
#4.Random component or irregular component:- this is the variation that is realastic or resedual and donot fall under any of the above 3components
```

```python
import numpy as np
import pandas as pd

# Create a date range starting from '1/1/2018', with 100 periods, at 1-second intervals
rng = pd.date_range('1/1/2018', periods=100, freq="S")

# Print the generated date range
print(rng)

# Create a pandas Series with random integers between 0 and 200, using the date range as the index
ts = pd.Series(np.random.randint(0, 200, len(rng)), index=rng)

# Display the first 5 rows of the time series
ts.head()
```

```python
ts.resample('5min').sum()
```

```python
#localize a time series to the time zone
rng=pd.date_range("1/1/2024 00:00", periods=5, freq='h')
print(rng)
```

```python
ts=pd.Series(np.random.randn(len(rng)), rng)
ts_utc=ts.tz_localize('UTC')
ts_utc
```

```python
#Convert the time stamp to another zone
ts_utc.tz_convert('america/new_york')
```

**PLOTTING WITH PANDAS**

```python
import matplotlib.pyplot as plt
ts=pd.Series(np.random.randn(1000),index=pd.date_range('1/1/2020', periods=1000))
ts=ts.cumsum()
ts.plot()
```

```python
df=pd.DataFrame(np.random.randn(1000,4), index=ts.index, columns=['A','B','C','D'])
#print(df)
df=df.cumsum()
df.plot()
plt.legend(loc = 'best')
plt.title("time series data with random varaibles")
```

. #Bar Chart
Purpose:

Used to compare quantities across different categories.

Best for categorical data like product sales, survey responses, or regions.

```python
import matplotlib.pyplot as plt

data = {'months': ['jan', 'feb', 'march', 'april', 'may'], 'sales': [200, 210, 220, 230, 240]}

# Extracting months and sales from the data dictionary
months = data['months']
sales = data['sales']

# Plotting a bar chart
plt.bar(months, sales)

# Adding labels and title
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales')
plt.show()
```
